1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome(r'C:\Users\91797\Desktop\chromedriver.exe')
```

```python
#Getting URL
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
#creating the empty list
Name=[]
Rank=[]
Artist=[]
Upload_date=[]
Views=[]


#extracting the Names of the Video
try:
    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
    
    
#extracting the Ranks of the Video    
    
try:
    ranks=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranks:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")
    
    

#extracting the Artist of the Video    
    
try:
    Artists=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in Artists:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
time.sleep(5)
    
#extracting the Upload Date of the Video
    
try:
    Upload_dates=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in Upload_dates:
        Upload_date.append(i.text)
except NoSuchElementException:
    Upload_date.append("-")
time.sleep(5)
    

#extracting the Views of the Video
    
    
try:
    views=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in views:
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
time.sleep(5)    


#creating the DataFrame
Details=pd.DataFrame({})
Details['Name']=Name
Details['Rank']=Rank
Details['Artist']=Artist
Details['Upload_date']=Upload_date
Details['Views in Billion']= Views
Details

    
```

```python
driver = webdriver.Chrome(r'C:\Users\91797\Desktop\chromedriver.exe')
```

```python
url = 'https://github.com/' #getting the required url
driver.get(url)

time.sleep(2)
```

```python
# Getting explore button and clicking on it
explore = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details").click()
```

```python
#Selecting trending option
trend_url = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
urls = trend_url.get_attribute("href")
driver.get(urls)
```

```python
# making empty lists    
repo_urls=[]
Contributors_count=[]
Language_used=[]
repo_title=[]
repo_desc=[]

#scraping repositories urls
repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in repos:
    repo_urls.append(i.get_attribute('href'))
time.sleep(2) 

    
#scraping repositories title
repo_title=[]
try:
    repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
    for i in repos:
        repo_title.append(i.text)
except NoSuchElementException:#handling no such element exception
    repo_title.append('-')
time.sleep(3)
    

for i in repo_urls:
    driver.get(i)
    l=[]
    time.sleep(3)
    #scraping repositories discription
    try:
        repo = driver.find_element_by_xpath("//p[@class='col-9 color-fg-muted my-1 pr-4']")
        repo_desc.append(repo.text)
    except NoSuchElementException:
        repo_desc.append('-')
    
    #scraping contributors count
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors_count.append(count.text)
    except NoSuchElementException:#handling no such element exception
        Contributors_count.append('No details available')
    time.sleep(2)
    
     #scraping languages used   
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:    
        for i in languages:
            l.append(i.text)
    else:
        l.append('No languages used')
    Language_used.append(l)   
    
time.sleep(2)   
#creating dataframe
Repositories=pd.DataFrame({})
Repositories['Title']=repo_title
Repositories['Description']=repo_desc
Repositories['Contributors_count']=Contributors_count
Repositories['Language_used']=Language_used
Repositories
```

A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre

```python
driver = webdriver.Chrome(r'C:\Users\91797\Desktop\chromedriver.exe')
```

```python
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
driver.get(url)
time.sleep(2)
```

```python
#scrapping book names
book_name=[]
author_name=[]
Sale=[]
publisher=[]
genre=[]


#scrapping book names
try:
    names=driver.find_elements_by_xpath("//table[@class='in-article sortable']/tbody/tr/td[2]")
    for i in names:
        book_name.append(i.text)
except NoSuchElementException:              #handling no such element exception
    book_name.append('-')
time.sleep(2)

#scrapping author names
try:
    names=driver.find_elements_by_xpath("//table[@class='in-article sortable']/tbody/tr/td[3]")
    for i in names:
        author_name.append(i.text)
except NoSuchElementException:#handling no such element exception
    author_name.append('-')
time.sleep(2)


#scrapping Sale
try:
    sale=driver.find_elements_by_xpath("//table[@class='in-article sortable']/tbody/tr/td[4]")
    for i in sale:
        Sale.append(i.text)
except NoSuchElementException:#handling no such element exception
    Sale.append('-')
time.sleep(2)    


#scrapping publisher
try:
    publishers=driver.find_elements_by_xpath("//table[@class='in-article sortable']/tbody/tr/td[5]")
    for i in publishers:
        publisher.append(i.text)
except NoSuchElementException:#handling no such element exception
    publisher.append('-')
time.sleep(2)

#scrapping genre

try:
    genres=driver.find_elements_by_xpath("//table[@class='in-article sortable']/tbody/tr/td[6]")
    for i in genres:
        genre.append(i.text)
except NoSuchElementException:#handling no such element exception
    genre.append('-')
time.sleep(2)



#Creating the DataFrame
Novels=pd.DataFrame({})
Novels['Book_name']= book_name
Novels['Author_name']=author_name
Novels['Volume sold']=Sale
Novels['Publisher']=publisher
Novels['Genre']=genre
Novels
```

# 9. Scrape the details most watched tv series of all time from imdb.com.

Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:

A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
driver = webdriver.Chrome(r'C:\Users\91797\Desktop\chromedriver.exe')
```

```python
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
#creating empty list
name=[]
years=[]
genre=[]
run_time=[]
rating=[]
votes=[]


# scraping name of webseries
try:
    names=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
    for i in names:
        name.append(i.text)
except NosuchElementException:#handling no such element exception
    name.append('-')
time.sleep(2)


# scraping year of release
try:
    span=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]")
    for i in span:
        years.append(i.text)
except NosuchElementException:#handling no such element exception
    years.append('-')
time.sleep(2)


# scraping genre
try:
    genres=driver.find_elements_by_xpath("//span[@class='genre']")
    for i in genres:
        genre.append(i.text)
except NoSuchElementException:#handling no such element exception
    genre.append('-')
time.sleep(2)


# scraping run_time
try:
    info=driver.find_elements_by_xpath("//span[@class='runtime']")
    for i in info:
        run_time.append(i.text)
except NoSuchElementException:#handling no such element exception
    run_time.append('-')
time.sleep(2)


# scraping rating
try:
    info=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")
    for i in info:
        rating.append(i.text)
except NoSuchElementException:#handling no such element exception
    rating.append('-')
time.sleep(2)

# scraping votes
try:
    info=driver.find_elements_by_xpath("//span[@name='nv']")
    for i in info:
        votes.append(i.text)
except NoSuchElementException:#handling no such element exception
    votes.append('-')
time.sleep(2)



#creating DataFrame
Series=pd.DataFrame({})
Series['Name']=name
Series['Years']=years
Series['Genre']=genre
Series['Run_time']=run_time
Series['Rating']=rating
Series['Votes']=votes
Series
```

```python
url = 'https://archive.ics.uci.edu/ml/index.php'
driver.get(url)
time.sleep(2)
```

```python
#clicking show all datasets link
show=driver.find_element_by_xpath("//a[@href='datasets.php']")
show.click()
time.sleep(2)
```

```python
#creating Emplty list
name=[]
data_type=[]
task=[]
attr_type=[]
no_of_instances=[]
no_of_attributes=[]
year=[]


#scraping dataset names
try:
    names=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in names:
        name.append(i.text)
except NoSuchElementException:#handling no such element exception
    name.append('-')
time.sleep(2)

#scraping data type
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
    for i in info[1:]:
         data_type.append(i.text)
except NoSuchElementException:
    data_type.append("-")
time.sleep(2)


#scraping dataset names
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
    for i in info[1:]:
         task.append(i.text)
except NoSuchElementException:
         task.append("-")
time.sleep(2)


#scraping dataset names
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
    for i in info[1:]:
         attr_type.append(i.text)
except NoSuchElementException:
    attr_type.append("-")
time.sleep(2)


#scraping dataset names
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
    for i in info[1:]:
         no_of_instances.append(i.text)
except NoSuchElementException:
    no_of_instances.append("-")
time.sleep(2)


#scraping dataset names
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
    for i in info[1:]:
         no_of_attributes.append(i.text)
except NoSuchElementException:
    no_of_attributes.append("-")
time.sleep(2)


#scraping dataset names
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
    for i in info[1:]:
         year.append(i.text)
except NoSuchElementException:
    year.append("-")
time.sleep(2)



#creating the DataFrame
Repositories=pd.DataFrame({})
Repositories['Name']=name
Repositories['Data_type']=data_type
Repositories['Task']=task
Repositories['Attribute_type']=attr_type
Repositories['No_of_instances']=no_of_instances
Repositories['No_of_attributes']=no_of_attributes
Repositories['Year']=year
Repositories




```
